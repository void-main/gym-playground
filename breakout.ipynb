```python
import gym
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
import time

%matplotlib inline
```

```python
env = gym.make('Breakout-v0')

print(env.observation_space.shape)
print(env.action_space.n)
```

```python
done = False
ob = env.reset()

while not done:
    act = random.randrange(4)
    next_ob, reward, done, _ = env.step(act)
    
    m = cv2.max(ob, next_ob)
    img_yuv = cv2.cvtColor(m, cv2.COLOR_RGB2YUV)
    y, _, _ = cv2.split(img_yuv)
    o = cv2.resize(y, (84, 84))
    print(o.shape)
    
    ob = next_ob
    plt.imshow(o)
```
